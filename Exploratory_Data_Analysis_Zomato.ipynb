```python
sns.countplot(df['listed_rest_type'])
fig=plt.gcf()
fig.set_size_inches(10,10)
```

```python
sns.countplot(df['cost'])
sns.countplot(df['cost']).set_xticklabels(sns.countplot(df['cost']).get_xticklabels(), rotation=90, ha="right")
fig = plt.gcf()
fig.set_size_inches(13,13)
plt.title('Location')
```

```python
sns.countplot(df['rest_type'])
sns.countplot(df['rest_type']).set_xticklabels(sns.countplot(df['rest_type']).get_xticklabels(), rotation=90, ha="right")
fig = plt.gcf()
fig.set_size_inches(15,15)
plt.title('Restuarant Type')
```

```python
#Types of Services
sns.countplot(df['online_order'])
sns.countplot(df['online_order']).set_xticklabels(sns.countplot(df['online_order']).get_xticklabels(), rotation=90, ha="right")
fig = plt.gcf()
fig.set_size_inches(15,15)
plt.title('Online Order or Not')
```

```python
sns.countplot(df['listed_rest_type'])
sns.countplot(df['listed_rest_type']).set_xticklabels(sns.countplot(df['listed_rest_type']).get_xticklabels(), rotation=90, ha="right")
fig = plt.gcf()
fig.set_size_inches(25,20)
plt.title('Restaurants Listed in (City)')
```

```python
#Types of Services
sns.countplot(df['rate'])
sns.countplot(df['rate']).set_xticklabels(sns.countplot(df['rate']).get_xticklabels(), rotation=90, ha="right")
fig = plt.gcf()
fig.set_size_inches(25,15)
plt.title('Ratings')
```
